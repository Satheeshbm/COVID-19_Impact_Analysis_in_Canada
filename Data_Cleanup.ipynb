```python
#Import libraries
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
#Assign paths
quebec_covid19_path = Path('Resources/COVID-19_Hospitalizations/Quebec_COVID-19_Hospitalizations.csv')
ontario_covid19_path = Path('Resources/COVID-19_Hospitalizations/Ontario_COVID-19_Hospitalizations.csv')
alberta_covid19_path = Path('Resources/COVID-19_Hospitalizations/Alberta_COVID-19_hospitalizations.csv')
saskatchewan_covid19_path = Path('Resources/COVID-19_Hospitalizations/Saskatchewan__COVID-19_Hospitalizations.csv')
manitoba_covid19_path = Path('Resources/COVID-19_Hospitalizations/Manitoba_COVID-19_hospitalizations.csv')
```

```python
#Read CSV files through Pandas
quebec_covid19_df = pd.read_csv(quebec_covid19_path, index_col='Date', infer_datetime_format=True, parse_dates=True)
ontario_covid19_df = pd.read_csv(ontario_covid19_path, index_col='date', infer_datetime_format=True, parse_dates=True))
alberta_covid19_df = pd.read_csv(alberta_covid19_path, index_col='Date reported to Alberta Health', infer_datetime_format=True, parse_dates=True))
saskatchewan_covid19_df = pd.read_csv(saskatchewan_covid19_path, index_col='Date', infer_datetime_format=True, parse_dates=True))
manitoba_covid19_df = pd.read_csv(manitoba_covid19_path, index_col='Date', infer_datetime_format=True, parse_dates=True))
```

```python
quebec_covid19_df.tail()
```

```python
#Drop unnecessary columns + reverse row order to match other COVID-19 datasets
quebec_covid19_df.drop(quebec_covid19_df.iloc[:, 0:38], inplace=True, axis=1)
quebec_covid19_df = quebec_covid19_df.iloc[::-1]
```

```python
quebec_covid19_df.head()
```

```python
#Sum total values per row to capture daily total hospitalization count
quebec_covid19_hospitalizations = quebec_covid19_df.sum(axis=1)
quebec_covid19_hospitalizations
```

```python
quebec_covid19_hospitalizations.columns = ['Quebec_Count']
#Matthew suggested adding a column at the end of the CSV with the aggregated count and add the sum numbers there. That's the method he's aware of
```
